```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

duracao_ep = ctrl.Antecedent(np.arange(0, 101, 1), 'duracao_ep')
idade = ctrl.Antecedent(np.arange(0, 101, 1), 'idade')
res = ctrl.Consequent(np.arange(0, 101, 1), 'res')

duracao_ep['baixo'] = fuzz.trapmf(duracao_ep.universe, [0, 0, 20, 25])
duracao_ep['medio'] = fuzz.trapmf(duracao_ep.universe, [20, 30, 35, 45])
duracao_ep['longo'] = fuzz.trapmf(duracao_ep.universe, [35, 45, 55, 60])
duracao_ep['muito longo'] = fuzz.trapmf(duracao_ep.universe, [55, 60, 100, 100])

idade['baby'] = fuzz.trapmf(idade.universe, [0, 0, 7, 9])
idade['crianca'] = fuzz.trapmf(idade.universe, [6, 8, 11, 13])
idade['pre-adolescente'] = fuzz.trapmf(idade.universe, [12, 13, 14, 15])
idade['adolescente'] = fuzz.trapmf(idade.universe, [14, 15, 19, 21])
idade['adulto 1'] = fuzz.trapmf(idade.universe, [19, 21, 38, 42])
idade['adulto 2'] = fuzz.trapmf(idade.universe, [38, 42, 58, 62])
idade['idoso'] = fuzz.trapmf(idade.universe, [58, 62, 100, 100])

res['nada'] = fuzz.trapmf(res.universe, [0, 0, 0, 0])
res['pouco'] = fuzz.trapmf(res.universe, [1, 1, 35, 45])
res['medio'] = fuzz.trapmf(res.universe, [35, 45, 60, 70])
res['muito'] = fuzz.trapmf(res.universe, [60, 70, 85, 90])
res['muito alto'] = fuzz.trapmf(res.universe, [85, 90, 100, 100])
```

```python
# Friends - 22min | Adolescente ou Adulto 1



r1 = ctrl.Rule(idade['baby'] & duracao_ep['baixo'], res['nada'])
r23 = ctrl.Rule(idade['baby'] & duracao_ep['medio'], res['nada'])
r24 = ctrl.Rule(idade['baby'] & duracao_ep['longo'], res['nada'])
r25 = ctrl.Rule(idade['baby'] & duracao_ep['muito longo'], res['nada'])

r2 = ctrl.Rule(idade['crianca'] & duracao_ep['baixo'], res['pouco'])
r26 = ctrl.Rule(idade['crianca'] & duracao_ep['medio'], res['pouco'])
r27 = ctrl.Rule(idade['crianca'] & duracao_ep['longo'], res['pouco'])
r28 = ctrl.Rule(idade['crianca'] & duracao_ep['muito longo'], res['pouco'])

r3 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['baixo'], res['muito'])
r4 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['medio'], res['medio'])
r5 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['longo'], res['pouco'])
r6 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['muito longo'], res['nada'])

r7 = ctrl.Rule(idade['adolescente'] & duracao_ep['baixo'], res['muito alto'])
r8 = ctrl.Rule(idade['adolescente'] & duracao_ep['medio'], res['muito'])
r9 = ctrl.Rule(idade['adolescente'] & duracao_ep['longo'], res['medio'])
r10 = ctrl.Rule(idade['adolescente'] & duracao_ep['muito longo'], res['pouco'])

r11 = ctrl.Rule(idade['adulto 1'] & duracao_ep['baixo'], res['muito alto'])
r12 = ctrl.Rule(idade['adulto 1'] & duracao_ep['medio'], res['muito'])
r13 = ctrl.Rule(idade['adulto 1'] & duracao_ep['longo'], res['medio'])
r14 = ctrl.Rule(idade['adulto 1'] & duracao_ep['muito longo'], res['pouco'])

r15 = ctrl.Rule(idade['adulto 2'] & duracao_ep['baixo'], res['muito'])
r16 = ctrl.Rule(idade['adulto 2'] & duracao_ep['medio'], res['medio'])
r17 = ctrl.Rule(idade['adulto 2'] & duracao_ep['longo'], res['pouco'])
r18 = ctrl.Rule(idade['adulto 2'] & duracao_ep['muito longo'], res['nada'])

r19 = ctrl.Rule(idade['idoso'] & duracao_ep['baixo'], res['medio'])
r20 = ctrl.Rule(idade['idoso'] & duracao_ep['medio'], res['pouco'])
r21 = ctrl.Rule(idade['idoso'] & duracao_ep['longo'], res['nada'])
r22 = ctrl.Rule(idade['idoso'] & duracao_ep['muito longo'], res['nada'])

regrasFriends = [r1, r2, r3, r4, r5, r6, r7, r8, r9, r10, r11, r12, r13, r14, r15, r16, r17, r18, r19, r20, r21, r22, r23, r24, r25, r26, r27, r28]
```

```python
# Vikings - 47min | Adolescente, Adulto 1 ou Adulto 2



v1 = ctrl.Rule(idade['baby'] & duracao_ep['baixo'], res['nada'])
v23 = ctrl.Rule(idade['baby'] & duracao_ep['medio'], res['nada'])
v24 = ctrl.Rule(idade['baby'] & duracao_ep['longo'], res['nada'])
v25 = ctrl.Rule(idade['baby'] & duracao_ep['muito longo'], res['nada'])

v2 = ctrl.Rule(idade['crianca'] & duracao_ep['baixo'], res['nada'])
v26 = ctrl.Rule(idade['crianca'] & duracao_ep['medio'], res['nada'])
v27 = ctrl.Rule(idade['crianca'] & duracao_ep['longo'], res['pouco'])
v28 = ctrl.Rule(idade['crianca'] & duracao_ep['muito longo'], res['pouco'])

v3 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['baixo'], res['pouco'])
v4 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['medio'], res['medio'])
v5 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['longo'], res['muito'])
v6 = ctrl.Rule(idade['pre-adolescente'] & duracao_ep['muito longo'], res['medio'])

v7 = ctrl.Rule(idade['adolescente'] & duracao_ep['baixo'], res['pouco'])
v8 = ctrl.Rule(idade['adolescente'] & duracao_ep['medio'], res['muito'])
v9 = ctrl.Rule(idade['adolescente'] & duracao_ep['longo'], res['muito alto'])
v10 = ctrl.Rule(idade['adolescente'] & duracao_ep['muito longo'], res['muito'])

v11 = ctrl.Rule(idade['adulto 1'] & duracao_ep['baixo'], res['pouco'])
v12 = ctrl.Rule(idade['adulto 1'] & duracao_ep['medio'], res['muito'])
v13 = ctrl.Rule(idade['adulto 1'] & duracao_ep['longo'], res['muito alto'])
v14 = ctrl.Rule(idade['adulto 1'] & duracao_ep['muito longo'], res['muito'])

v15 = ctrl.Rule(idade['adulto 2'] & duracao_ep['baixo'], res['pouco'])
v16 = ctrl.Rule(idade['adulto 2'] & duracao_ep['medio'], res['muito'])
v17 = ctrl.Rule(idade['adulto 2'] & duracao_ep['longo'], res['muito alto'])
v18 = ctrl.Rule(idade['adulto 2'] & duracao_ep['muito longo'], res['medio'])

v19 = ctrl.Rule(idade['idoso'] & duracao_ep['baixo'], res['pouco'])
v20 = ctrl.Rule(idade['idoso'] & duracao_ep['medio'], res['medio'])
v21 = ctrl.Rule(idade['idoso'] & duracao_ep['longo'], res['muito'])
v22 = ctrl.Rule(idade['idoso'] & duracao_ep['muito longo'], res['medio'])

regrasVikings = [v1, v2, v3, v4, v5, v6, v7, v8, v9, v10, v11, v12, v13, v14, v15, v16, v17, v18, v19, v20, v21, v22, v23, v24, v25, v26, v27, v28]
```

```python
idade_ = 40
duracaoep_ = 24

friends_ctrl = ctrl.ControlSystem(regrasFriends)
friends = ctrl.ControlSystemSimulation(friends_ctrl)

#setando informaes do usurio de usuario
friends.input['idade'] = idade_
friends.input['duracao_ep'] = duracaoep_

friends.compute()

print ('Recomendao da srie (%):', friends.output['res'])
res.view(sim=friends)

#GRAFICO RESULTANTE DA SRIE FRIENDS
```
